# Scrap name of different types of sugar and its variants on wikipedia

```python
from bs4 import BeautifulSoup
import requests

page = requests.get('https://en.wikipedia.org/wiki/List_of_sugars')
soup = BeautifulSoup(page.text, 'html.parser')
```

## Extract relevant items

```python
soup
```

```python
sugar_table = soup.find_all('ul')[1]
sugar_table
```

```python
sugar_row = sugar_table.find_all('li')
```

```python
sugar_row = list(sugar_row)
sugar_row
```

```python
for i in range(len(sugar_row)):
    sugar_row[i] = str(sugar_row[i])
```

```python
# 2 types, one with hyper link and one without
import re

sugar_name = []

for i in range(len(sugar_row)):
    sl = re.findall(r'title="[\w\W]+?"',sugar_row[i])
    for s in sl:
        sugar_name.append(s[7:-1])
    
    if len(sl) ==0:
        sl = re.findall(r'>[\w\W]+?<', sugar_row[i])
        sugar_name.append(sl[0][1:-1])
```

```python
sugar_name
```

### Standardise the names

```python
sugar_name.index('Corn sugar (disambiguation)')
```

```python
sugar_name[19] = 'Corn sugar'
```

```python
sugar_name.index('Fruit juice, Fruit juice concentrate')
```

```python
sugar_name[40] = 'Fruit juice'
sugar_name.append('Fruit juice concentrate')
```

```python
sugar_name.index("Refiner's sugar (page does not exist)")
```

```python
sugar_name[87] = "Refiner's sugar"
```

```python
sugar_name.index("Yellow sugar (page does not exist)")
```

```python
sugar_name[109]  = 'Yellow sugar'
```

### Lower case and remove duplicates

```python
sugar_name = list(set(sugar_name))

for i in range(len(sugar_name)):
    sugar_name[i] = sugar_name[i].lower()
```

```python
sugar_name
```

# find number of types of sugar in the product

```python
import pandas as pd
branded_food = pd.read_csv('branded_food.csv')
```

```python
branded_food.head()
```

```python
branded_food.shape
```

```python
sugar_type = set(sugar_name)
```

```python
# replace NA in ingredients
branded_food['ingredients'].fillna('NA', inplace = True)
# change all ingredients to lower
branded_food['ingredients'] = branded_food['ingredients'].str.lower()
```

```python
sugar_count = []
sm = []
# iteratate through dataframe to find match
for i in range(len(branded_food)):
    sugar = []
    # using comma as delimiter
    ingredients = branded_food.loc[i, 'ingredients'][:-1].split(', ')
    
    for item in ingredients:
        if item in sugar_type:
            sugar.append(item)
    
    sugar_count.append(len(sugar))
    sm.append(sugar)
```

```python
branded_food['sugar_name'] = sm
branded_food['sugar_count'] = sugar_count
```

```python
branded_food[branded_food['sugar_count'] > 0].describe()
```

```python
len(branded_food[branded_food['sugar_count'] > 0]) / len(branded_food)
```

```python
branded_food[branded_food['sugar_count']  == 21]
```

- Over half of the package food contains added sugar.
- The average number of sugar types in the sample packaged food is 2. The product that contains the highest number of sugar types comes from the dairy category with a count of 21.

```python
branded_food[branded_food['sugar_count'] > 0].groupby('branded_food_category').mean().sort_values('sugar_count', ascending = False).head(10)
```

Some of the categories that include the most sugar and its variants are:
- pies, pastries
- cereal / muesli bars
- meat substitutes

# Scrap salad dressing nutritional label from Praise

### Coleslaw dressing

```python
from bs4 import BeautifulSoup
import requests

page = requests.get('http://www.praise.com.au/our-products/praise-coleslaw-dressing-330ml/')
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
html_string = str(soup)
df = pd.read_html(html_string)
```

```python
normal_coleslaw = df[0]
normal_coleslaw
```

```python
page = requests.get('http://www.praise.com.au/our-products/praise-coleslaw-fat-free-dressing-330ml/')
soup = BeautifulSoup(page.text, 'html.parser')
html_string = str(soup)
df = pd.read_html(html_string)
fat_free_coleslaw = df[0]
fat_free_coleslaw
```

```python
coles_law_comb = pd.concat([normal_coleslaw, fat_free_coleslaw], axis = 1)
coles_law_comb.columns = ['nutrients','serving_normal', 'd1','per_100g_normal','d2','serving_fat_free','d4','per_100g_fat_free']
coles_law_comb.drop(['d1','d2','d4'], axis = 1, inplace= True)
coles_law_comb
```

```python
float(coles_law_comb.loc[5,'per_100g_fat_free']) / float(coles_law_comb.loc[5,'per_100g_normal'])-1 
```
